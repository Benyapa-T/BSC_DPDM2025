เนื่องจากข้อมูลมี Column 'measure_datetime' อยู่แล้ว สามารถทำการนำข้อมูลมาต่อกันในแนวแกน Y (Y-Data Integration) ได้เลย

```python
path = '/content/drive/MyDrive/BSC DPDM/Midterm/BSC DPDM midterm 2025/Input - (Operation Daily rain from API_ 2025)'
list_month = ['2025-07','2025-08','2025-09','2025-10','2025-11','2025-12']
```

```python
#Prompt: ต้องการดึงไฟล์ทั้งหมด ที่อยู่ใน path ออกมาจากโฟล์เดอร์แต่ละโฟล์เดอร์ list_month ทั้งหมดออกมาต่อกันเป็นไฟล์เดียว
all_df = []

for month in list_month:
    folder_path = os.path.join(path, month)

    if os.path.exists(folder_path):
        files = os.listdir(folder_path)

        for file in files:
            # เพิ่มเงื่อนไขตรวจสอบชื่อไฟล์ขึ้นต้นด้วย 'HII_daily_rainfall' และนามสกุล .csv
            if file.startswith('HII_daily_rainfall') and file.endswith('.csv'):
                file_path = os.path.join(folder_path, file)
                df = pd.read_csv(file_path)
                all_df.append(df)

        print(f"✅ โหลดไฟล์จากเดือน {month} เรียบร้อยแล้ว")

# รวมไฟล์
if all_df:
    input_df = pd.concat(all_df, ignore_index=True)
    print(f"\n✨ รวมไฟล์เสร็จสิ้น! พบข้อมูลทั้งหมด {len(input_df)} แถว")
else:
    print("\n⚠️ ไม่พบไฟล์ที่ตรงตามเงื่อนไข")

input_df
```

#4 คัดสถานีที่มีข้อมูลสูญหาย ≥20% ออก แสดงรายชื่อที่ถูกคัดออกและจำนวนที่เหลือ

## คัดสถานีที่มีข้อมูลสูญหาย ≥20% ออก

```python
input_df_drop.isnull().sum()
```

```python
import pandas as pd

# 1. สร้างตารางสรุปโดย Group ตาม tele_station_oldcode
# เราจะนับจำนวนแถวทั้งหมด (Total) และจำนวนค่าว่าง (Missing)
summary_df = input_df_drop.groupby('tele_station_oldcode').agg(
    total_rows=('tele_station_oldcode', 'count'),
    missing_rows=('tele_station_id', lambda x: x.isnull().sum()) # เปลี่ยน 'rainfall_value' เป็นชื่อ column ที่ต้องการเช็ค
).reset_index()

# 2. คำนวณเปอร์เซ็นต์ Missing
summary_df['missing_percentage'] = (summary_df['missing_rows'] / summary_df['total_rows']) * 100

# 3. เรียงลำดับจากสถานีที่ Missing เยอะที่สุดไปน้อยที่สุด
summary_df = summary_df.sort_values(by='missing_percentage', ascending=False)

# แสดงผล 10 อันดับแรก
print("ตารางสรุปข้อมูลแยกตามรหัสสถานี (10 อันดับแรกที่มี Missing สูงสุด):")
print(summary_df.head(10).to_string(index=False))

# 4. สรุปภาพรวม
print(f"\nจำนวนสถานีทั้งหมด: {len(summary_df)} สถานี")
```

เนื่องจากไม่มีสถานีที่ข้อมูลสูญหาย ≥20% จึงไม่ได้ทำการคัดออก

สร้าง column month year

```python
# Prompt : หา เดือน ปี จาก column measure_datetime ในตาราง input_df_drop ที่มีการบันทึกข้อมูลเป็น 2025-07-09 ปี-เดือน-วัน โดยใช้ฟังก์ชั่น split

# สร้างคอลัมน์ 'month'
input_df_drop['month'] = input_df_drop['measure_datetime'].astype(str).str.split('-').str[1]

# สร้างคอลัมน์ 'year'
input_df_drop['year'] = input_df_drop['measure_datetime'].astype(str).str.split('-').str[0]
```

```python
input_df_drop.head()
```

```python
# Prompt: สร้างตารางใหม่ ชื่อ Output แล้วดึงข้อมูล tele_station_oldcode จากตาราง input_df_drop มาสร้างใหม่ชื่อว่า Node โดยข้อมูลที่ดึงมาต้องไม่ซ้ำกัน
Output = pd.DataFrame({
    'Node': input_df_drop['tele_station_oldcode'],
    'year': input_df_drop['year'],
    'month': input_df_drop['month']
})
Output.drop_duplicates()
```

```python
rain_sum = (
    input_df_drop
    .groupby(['year', 'month'])['rainfall_value']
    .sum()
    .reset_index()
)

rain_sum = rain_sum.rename(columns={'rainfall_value': 'rain'})
Output = Output.merge(rain_sum, on=['year','month'], how='left')
```

```python
Output = Output.sort_values(by='Node').reset_index(drop=True)
Output
```
